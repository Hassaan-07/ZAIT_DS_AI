```python
import os #isy islye import kya h k directory se related hony wly operation perform krsku.

'''
directory structure:
    useraccounts > cnic1.txt , cnic2.txt , .... , cnicN.txt ------> record each user transaction
    users.txt ------> carry user credentials like name , cnic , password to verify existin user of our bank
    
'''

# global variables
account = {
    "name":None,
    "balance":0.0,
    "transactions":[]
}

# ye function ye check krta h ke ye cnic ka user database m hai ya nhi agr hota tu signup pr btadeta k ye already registered user h tu signup mt kro.
def cnic_check(cnic):
    files = os.listdir('.')
    for file in files:
        if file == "users.txt":
            with open("users.txt") as users:
                for user in users:
                    user_data = user.split(',')
                    if cnic == user_data[1]:
                        return True
                return False
                    
    return False

# ye function cnic ki validation krta h kyuke hmry pakistani cnic mai 13 digits hty hai.
def cnic_validation(cnic):      
    return cnic.isnumeric() and len(cnic)==13

# ye login krta h user ko user.txt ki file ko check krke agr user usmai registered hai tbi login hota wrna nhi.
def login():
    while True:
        cnic = input("Enter cnic: ")
        if cnic_validation(cnic):
            upass = input("Enter password: ")
            break
        else:
            print("Please enter a valid cnic...")

    
    try:
        with open("users.txt") as users:
            for user in users:
                uacc = user.replace("\n","").split(',') 
                if cnic == uacc[1] and upass == uacc[2]:
                    account["name"] = uacc[0]
                    print("Login sucessfully")
                    if after_login(cnic) == None:
                        break

            else:
                print("User not exist try to register first")
        
    except:
        print("this user is not exist try to register user first.")
        
        



'''
ye function new user account create krta h sth mai check b krta h ke new user jb apna cnic enter krta h tu usy validate krke check krta h ke ye kahi 
pehly se registered user tu nhi wrna usy login ka kheta h else signup k other fields ki trf bhj deta h.

'''
def createAccount():
    while True:
        cnic = input("Enter cnic (13 digit number): ")
        if cnic_validation(cnic):
            if cnic_check(cnic):
                print("this user is already registered..")
                break
            else:
                uname = input("Enter username: ")
                upass = input("Enter password")
                with open("users.txt","a+") as ur:
                    user = ",".join([uname , cnic , upass])
                    ur.write(user+"\n")
                print(f"Account created sucessfully for {uname}..")
                f = user_file_handler(cnic,'w')
                f.write("0.0\n\n")
                f.close()
                break
        else:
            print("enter a valid cnic number..")
    

'''
ye function login sucesfull hone k baaad bank account mai hony wly operations k bry mai user ko display dikha kr us se input lekr user ko particular
bank account k operation pr bhjta h 
'''

def after_login(cnic):
    while True:
        print(f'''
                1. Deposit
                2. Withdraw
                3. Balance
                4. Statement
                5. Back
        ''')
        try:
            key = int(input("Enter key: "))
            get_data(cnic)
            if key==1:
                deposit(cnic)
            elif key==2:
                withdraw(cnic)
            elif key==3:
                checkBalance()
            elif key==4:
                printStatement()
            elif key==5:
                return None
            else:
                print("Enter a valid key")
                
        except:
            print("Enter a valid key")
    

'''
ye function useraccount k folder k andr ar 1 user ki transactions ki files bnata h or ar 1 user ki transaction file us user ke unique cnic number se bn
kr ar user ki transactions us user k separate file m bnti h or ye file handler ka object return krta h.
'''
def user_file_handler(cnic,mode):
    file = f"useraccounts/{cnic}.txt"
    f = open(file,mode)
    return f


'''
ye function logged in user ki transaction file se uske data ko utha kr account named dictionary based datastructure mai store krta hai.
'''
def get_data(cnic):
    account["transactions"] = []
    f = user_file_handler(cnic,'r')
    data = f.readlines()
    f.close()
    # account['cnic'] =  cnic
    account["balance"] = float(data[0].replace("\n",""))
    for line in data[1::]:
        account["transactions"].append(line)
    return None


'''
ye function deposit or withdraw k amount ko check krta h ke user n positive amount value dli h ya negative
'''
def input_check():
    while True:
        try:
            amount = float(input("Enter amount: "))
            if amount >= 0:
                return amount
            else:
                return -1
        except:
            print("enter a valid amount")

'''
ye function user k transaction file ko new krke bna kr uski purani transaction ko account dictionary datastructrue se utha kr new file m daal kr 
uske sth deposit se related data ko b add krta h or balance ko b increase krta hai.
'''

def deposit(cnic):
    print("Deposit Panel")
    amount = input_check()
    if amount != -1:
        # get_data(cnic) ye yha krny se acha h logged in pr krleta hoo wha se 1 dfa call hga wrma ar operation pr call krke global datastructure ko update krna hga
        account["balance"] += amount
        f = user_file_handler(cnic,"w")
        f.write(str(account["balance"])+"\n")
        for line in account["transactions"]:
            f.write(line)
        text = f"- Deposit: {amount}. New balance: {account["balance"]}\n"
        f.write(text)
        f.close()
        print(f'''
           {account['name']} Deposited: {amount}. New balance: {account["balance"]}
        ''')
    else:
        print("Deposit amount is negative..")
        
'''
ye function user k transaction file ko new krke bna kr uski purani transaction ko account dictionary datastructrue se utha kr new file m daal kr 
uske sth withdraw se related data ko b add krta h or balance ko reduce krta hai withdraw ki amount ko kaat kr.
'''

def withdraw(cnic):
    print("Withdraw Panel")
    amount = input_check()
    if amount != -1:
        if amount <= account["balance"]:
            account["balance"] -= amount
            f = user_file_handler(cnic,"w")
            f.write(str(account["balance"])+"\n")
            for line in account["transactions"]:
                f.write(line)
            text = f"- Withdrawal: {amount}. New balance: {account["balance"]}\n"
            f.write(text)
            f.close()
            print(f'''
            {account['name']} Withdrew: {amount}. New balance: {account["balance"]}
            ''')
        else:
            print("withdraw can't be proceed becasue your withdraw amount is higher than current balance")
    else:
        print("Deposit amount is negative..")


'''
ye logged in user k account ki balance ki information provide krta h jo account k useraccount datastructure mai hti h.
'''
def checkBalance():
    print(f'''
        {account['name']} Current Balance: {account["balance"]}
    ''')

'''
ye logged in user k account statement ko show krta hai.    
'''
def printStatement():
    length = len(account["name"]) + 42
    print(f'''
            {"-"*10}ACCOUNT STATEMENT FOR {account['name']}{"-"*10}
            Account Balance: {account["balance"]}''')
    if len(account["transactions"])>1:
        for transaction in account["transactions"]:
            print(f"\t\t{transaction.replace("\n","")}")
    else:
        print(f'            {account["name"]} not done any transaction yet')
    print(f"            {"-"*length}")
        
    
'''
ye main ka function hai jo start hoty hi useraccounts naam se folder bnadeta h jha sb user ki transaction files user k cnic k naam se svae hti hti hai
or uske ilawa yeh hamary bank project ke flow ko start krta hai.
'''

def main():
    try:
        os.mkdir("useraccounts")
    except:
        print(end='')
    while True:
        print(f'''
                    {'-'*20}WELCOME TO PYTHON BANK{'-'*20}
    
                    1- Create Bank Account
                    2- Login Account
                    3- Exit
                    
        ''')
        try:
            user_key = int(input("Enter a key"))
            if user_key == 1:
                createAccount()
            elif user_key == 2:
                login()
            elif user_key == 3:
                print("Thank you for using Python Bank.")
                break
            else:
                print("Enter a valid key")
        except ValueError:
            print("Enter a valid key")
            
    
    
```

```python
# program starts
main()
```
